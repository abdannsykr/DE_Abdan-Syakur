```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("dataset_penjualan.csv")
df
```

```python
print("\nStatistik deskriptif:")
print(df.describe())
```

```python
print("\nJumlah nilai yang hilang:")
print(df.isnull().sum())
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Tanggal'], df['Jumlah_Penjualan'], marker='o', linestyle='-')
plt.title('Tren Penjualan Harian')
plt.xlabel('Tanggal')
plt.ylabel('Jumlah Penjualan')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
kategori_counts = df['Kategori_Produk'].value_counts()
plt.figure(figsize=(8, 6))
plt.bar(kategori_counts.index, kategori_counts.values)
plt.title('Distribusi Kategori Produk')
plt.xlabel('Kategori Produk')
plt.ylabel('Jumlah Penjualan')
plt.xticks(rotation=45)
plt.show()
```
