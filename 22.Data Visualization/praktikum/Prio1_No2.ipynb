```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('bakery.csv')
df
```

```python
nilai_transaksi_per_hari = df.groupby('transaction_date')['transaction_amount'].sum()
```

```python
jumlah_pesanan_per_roti = df.groupby('item_name').size()

roti_paling_banyak_dipesan = jumlah_pesanan_per_roti.idxmax()

kategori_roti = jumlah_pesanan_per_roti.index.tolist()
jumlah_pesan = jumlah_pesanan_per_roti.values.tolist()
```

```python
average_rating_per_day = df.groupby('transaction_date')['customer_rating'].mean()
```

### Jumlah nilai transaksi per hari.

```python
plt.figure(figsize=(15, 7))
plt.xlabel('Total Transaction Value')
plt.ylabel('Transaction Date')
plt.title('Total Transaction Value per Day')
plt.xticks(rotation=50)

plt.barh(nilai_transaksi_per_hari.index, nilai_transaksi_per_hari.values)
plt.show()
```

### Jenis roti yang paling banyak dipesan.

```python
plt.figure(figsize=(15, 7))
plt.xlabel('Item Name')
plt.ylabel('Count')
plt.title('Most Ordered Bread Types')
plt.xticks(rotation=80)

plt.bar(kategori_roti, jumlah_pesan) 
plt.show()
```

### Rata-rata kepuasan pelanggan (customer_rating) per hari.

```python
plt.figure(figsize=(15, 7))
plt.plot(average_rating_per_day.index, average_rating_per_day.values)

plt.xlabel('Transaction Date')
plt.ylabel('Average Customer Rating')
plt.title('Average Customer Rating per Day')

plt.xticks(rotation=50)

plt.show()
```
