```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('survey.csv')
df
```

### Pastikan data survei sudah diproses untuk menghindari data duplikat dan data dengan atribut yang tidak lengkap.

```python
df_cleaned = df.drop_duplicates().dropna()
print("Jumlah baris dan kolom setelah pembersihan data:", df_cleaned.shape)
```

### Persentase responden berdasarkan umur.

```python
age_counts = df_cleaned['age'].value_counts()
plt.figure(figsize=(15, 7))
plt.scatter(age_counts.index, age_counts.values, color='red')
plt.title('Distribusi Umur Responden')
plt.xlabel('Umur')
plt.ylabel('Jumlah Responden')
plt.grid(True)
plt.show()
```

### Bahasa pemrograman / framework yang paling banyak digunakan dalam pengembangan front end.

```python
frontend_counts = df_cleaned['favorite_front_end'].value_counts()

plt.figure(figsize=(15, 7))
frontend_counts.plot(kind='bar', color='green', label='Front-end')
plt.title('Bahasa pemrograman / framework yang paling banyak digunakan dalam pengembangan front end')
plt.xlabel('Bahasa Pemrograman/Framework')
plt.ylabel('Jumlah Responden')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

### Bahasa pemrograman / framework yang paling banyak digunakan dalam pengembangan back end.

```python
backend_counts = df_cleaned['favorite_back_end'].value_counts()

plt.figure(figsize=(15, 7))
backend_counts.plot(kind='bar', color='cornflowerblue', alpha=0.7, label='Back-end')
plt.title('Bahasa pemrograman / framework yang paling banyak digunakan dalam pengembangan back end.')
plt.xlabel('Bahasa Pemrograman/Framework')
plt.ylabel('Jumlah Responden')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

### Persentase bahasa pemrograman / framework yang dipilih dalam pengembangan front end.

```python
front_end_languages = df_cleaned['favorite_front_end'].dropna()

language_counts = front_end_languages.value_counts()

total_respondents = len(front_end_languages)

language_percentages = (language_counts / total_respondents) * 100

plt.figure(figsize=(15, 7))
language_percentages.sort_index().plot(kind='line', marker='o', color='purple')
plt.title('Persentase bahasa pemrograman / framework yang dipilih dalam pengembangan front end')
plt.xlabel('Bahasa Pemrograman/Framework')
plt.ylabel('Persentase (%)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

### Rentangan gaji per tahun yang paling banyak berdasarkan hasil survei.

```python
salary_counts = df_cleaned['salary_per_year_in_USD'].value_counts()

plt.figure(figsize=(15, 7))
plt.hist(df_cleaned['salary_per_year_in_USD'], bins=7, color='red', edgecolor='blue')
plt.title('Rentangan gaji per tahun yang paling banyak berdasarkan hasil survei')
plt.xlabel('Gaji Tahunan (USD)')
plt.ylabel('Jumlah Responden')
plt.show()
```
