```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('electronic_shop.csv')
df
```

### Membuat sebuah line plot untuk menggambarkan jumlah nilai dari transaksi per hari.

```python
result_transaction_date = df.groupby(['transaction_date']).sum(numeric_only=True)

transaction_dates = result_transaction_date.index.get_level_values(0).tolist()
transaction_amounts = result_transaction_date['transaction_amount']

plt.figure(figsize=(15, 7))

plt.xlabel('Transaction Date')
plt.ylabel('Transaction Amount')
plt.title('Transaction Amount by Date')

plt.plot(transaction_dates, transaction_amounts)

plt.show()
```

### Membuat sebuah bar plot untuk menggambarkan jumlah customer per hari.

```python
customer_per_day = df.groupby(['transaction_date']).count()

customer_dates = customer_per_day.index.get_level_values(0).tolist()
customer_amounts = customer_per_day['customer_name']

plt.figure(figsize=(15, 7))

plt.xlabel('Transaction Date')
plt.ylabel('Customer Amount')
plt.title('Customer Amount by Date')

plt.plot(customer_dates, customer_amounts)

plt.show()
```

### Membuat sebuah scatter plot untuk menggambarkan jumlah nilai dari transaksi per hari.

```python
result_transaction_date = df.groupby(['transaction_date']).sum()

transaction_dates = result_transaction_date.index.tolist()
transaction_amounts = result_transaction_date['transaction_amount']


plt.figure(figsize=(15, 7))
plt.xlabel('Transaction Date')
plt.ylabel('Transaction Amount')
plt.title('Transaction Amount by Date')

plt.scatter(transaction_dates, transaction_amounts)
plt.show()
```
