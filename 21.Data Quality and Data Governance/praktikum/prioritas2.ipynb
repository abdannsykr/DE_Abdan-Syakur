```python
import pandas as pd
```

```python
df = pd.read_csv('tx_data.csv')

df
```

```python
df['phone_number'] = df['phone_number'].apply(lambda x: str('+62' + str(x)))

df
```

```python
df['transaction_amount'] = df['transaction_amount'].apply(lambda x: 'Rp{:,}'.format(x))

df
```

```python
df = df[df['transaction_status'] == 'success']
```

```python
data_cleaning = df.copy()

print(data_cleaning)
```

```python
data_cleaning.to_csv('data_cleaning.csv', index=False)
```
